## [11-23-14] Results from analyzing data from 40-day-run run on 1st-article counters (hand-written notes for more details)

### Objective
This study stemmed from me observing that in the propaganda-plot, especially for the longer bars[set#>45(300cm)] , the time resolution was fluctuating (~5 ps) in a manner that we did not directly understand. The goal of this study was to rule out any statistical errors we had not possibly accounted for and therefore that did not propagate to the proganda-plot. The following sources of underestimated and/or unaccounted statistical errors could be possible:

1. __underestimated(usage of CSQ-WW in T-Fits) __:In the official data, the final fit of the T distribution is done using CSQ-WW(WWQ0), which in comparison with regular $\chi^{2}$(CSQ) Fit 
    * overestimates the time-resolution (especially for low statistics - n<= 500 - which is typical of data per Point in the 2-day-runs. As n increases, CSQ-WW and CSQ begin to agree more and more, though interestingly both differ from the true RMS. More details of this study can be seen in nb-comp-MLE-CSQ_MU-SG-TOF.)
    * underestimates the error on the time-resolution
2. __unaccounted (TW-parameters)__:Currently, the T distribution does not reflect the error in the estimation of the Time-Walk parameters. And not only that, there will be systematic effects in even estimatng the Time-Walk parameters because the entire process of their estimation involves a lot of Fitting and they are all done with the CSQ-W or MLE option (for 'tdc1-tw_1dimm'=WQO and 'tdc2_tw_2dimm'=MLNQR, 'lambda1d'=QL0)

### Data used
I collected roughly ~43 days worth of the data between 2nd Jul'14 and 13th Aug' 14 out of which I selected 40 days worth of data. The data was analyzed all together and in 2 day cycles, using the CSQ and CSQ-WW option to fit the final T histogram. 

__Note, the run of July-12-13 had a "problem" and therefore was removed from the analysis(see nb_40dr-2dc)__

### Results 

1. It can be seen from the resolution vs. bar-length plot that using the CSQ-WW option viz-a-viz CSQ in the estimation of the time-res from the final T Fits (n~500 which is typical of 2-day-runs)
    * significantly underestimates the stat-error per point(to the point that the res vs. bar-length plot looks to have fluctuations that are not statistical), which from the CSQ method, varies from 2.5ps in the middle to 5 ps at the edges(for a typical 2-day run).
    * overestimates the average time-resolution (this can be directly seen in the result plot in this study and/or in the individual files and in nb_299cm-bar_CSQ-MLE)
    * underestimates the error on the average time-resolution (open any of the result files to see this)
1. It can be directly seen that the average resolution obtained from each 2-day-run fluctuates within 2 ps; this is directly plotted in the following study.

Additionally:

1. The average resolution from each 2-day-run is systematically lower for the CSQ method (this is expected and can be seen in nb-comp-MLE-CSQ_MU-SG-TOF when n<=500 as is the case for a typical 2-day-run)
2. The average resolution from the Combined run (40-day-run) is very close for CSQ and MLE. Again, from nb-comp-MLE-CSQ_MU-SG-TOF, this can be expected since as 'n' grows, CSQ and MLE converge in their estimation of the RMS. (For completion, as per nb-comp-MLE-CSQ_MU-SG-TOF, I would have expected CSQ to be still smaller, but the difference at this point is getting very small) 

### Conclusion
The conclusion from this study is that there are under-represented(usage of CSQ-WW) and un-accounted (TW-pars) statistical errors.

1. __underestimated(usage of CSQ-WW in T-Fits)__:While these ultimately has minimal impact on the propaganda-plot (due to averaging-effect which reduces the propagated error), their impact can be seen in the resolution vs. bar length plot for each combination; when the CSQ option is used, the fluctuations are withing statistical error and with the CSQ-WW, the statistical errors are not large enough to enclose the fluctuations. 
1. __unaccounted(TW-parameters)__: Given that #1 does not produce a significant effect on the average resolution error, the fluctuation of the order of ~2 ps that can be seen in the average resolution obtained from various 2-day-runs can be attributed to any un-accounted for statistical errors, like that from TW-pars. However this fluctuation is of the order of the black-dot that represents each data point on the propaganda plot and therefore is not the cause of the larger fluctuations we see for the larger bars.

__This conclusion leads to the next hypothesis that the fluctuation could be due to there being much variability in the quality of the longer bars which finally manifests as the flucuations seen in the propaganda plot.__

```python
import rootnotes
datadir=os.path.join(os.environ['FTOF_DATADIR'],"analyzed","stat_err_ana")

def MyFn(s):
    #pos=s.find("July")
    if s.find("July")>=0:
        pos=s.find("July")
        date= s[pos+5:pos+7]
    elif s.find("August")>=0:
        pos=s.find("August")
        date= s[pos+7:pos+9]
    return int(date)
#! Get all 40dr-2dc- root files
fnl_Jul={}
#! First get all July files
for mthd in ['MLE','CSQ']:
    path="%s/40-day-run-2-day-cycles/%s"%(datadir,mthd)
    fnl_Jul[mthd]=[ f for f in os.listdir(path) if ( os.path.isfile(os.path.join(path,f)) and  
                                                os.path.join(path,f).find("July")>=0 and 
                                                os.path.join(path,f).find("July-12-13")<0)]
    fnl_Jul[mthd]=sorted(fnl_Jul[mthd],key=MyFn)
#! Now get August files
fnl_Aug={}
for mthd in ['MLE','CSQ']:
    path="%s/40-day-run-2-day-cycles/%s"%(datadir,mthd)
    fnl_Aug[mthd]=[ f for f in os.listdir(path) if ( os.path.isfile(os.path.join(path,f)) and  
                                                os.path.join(path,f).find("August")>=0 )]
    fnl_Aug[mthd]=sorted(fnl_Aug[mthd],key=MyFn)
#! Now put the two dicts together
fnl=dict(fnl_Jul)
fnl['MLE']+=fnl_Aug['MLE']
fnl['CSQ']+=fnl_Aug['CSQ']
#print fnl

#! Now open the ROOT files
fl={}
for mthd in ['MLE','CSQ']:
    path="%s/40-day-run-2-day-cycles/%s"%(datadir,mthd)
    fl[mthd] = [ ROOT.TFile("%s/%s"%(path,fn)) for fn in fnl[mthd] ]

#! Now get the combined 40dr
f_cmb={}
for mthd in ['MLE','CSQ']:
    path="%s/40-day-run/%s"%(datadir,mthd)
    f_cmb[mthd]=ROOT.TFile("%s/%s"%(path,"40-day-run_no-Jul-12-13_1st-article_321654.root"))
        


#! From fl,obtain h_avg_rsltnl and run_namel
h_avg_rsltnl={}
run_namel={}
for mthd in ['MLE','CSQ']: 
    run_namel[mthd]=[]
    h_avg_rsltnl[mthd]=[]
    for i,f in enumerate(fl[mthd]):
        name=f.GetName()[0:f.GetName().find("-cycle1")]
        if name.find("July")>=0:
            run_namel[mthd].append(name[name.find("July"):])
        elif name.find("August")>=0:
            run_namel[mthd].append(name[name.find("August"):])
        
        h=f.Get("fullbar_tw/avrg_resol_cont")
        h.SetTitle("%s:%s"%(h.GetTitle(),mthd))
        #h.SetMarkerColor(i+1)#ROOT.gROOT.ProcessLine
        #h.SetLineColor(i+1)
        if(i<=9):
            h.SetMarkerColor(ROOT.gROOT.ProcessLine("kViolet-%d"%i))#ROOT.gROOT.ProcessLine
            h.SetLineColor(ROOT.gROOT.ProcessLine("kViolet-%d"%i))
        else:
            h.SetMarkerColor(ROOT.gROOT.ProcessLine("kViolet+%d"%(i-9)))#ROOT.gROOT.ProcessLine
            h.SetLineColor(ROOT.gROOT.ProcessLine("kViolet+%d"%(i-9)))
        h.SetStats(0)
        h_avg_rsltnl[mthd].append(h)

#! From f_cmb,obtain h_avg_rsltn_cmb and run_name_cmb
h_avg_rsltn_cmb={}
run_name_cmb={}
for mthd in ['MLE','CSQ']: 
    run_name_cmb[mthd]="combined"
        
    h=f_cmb[mthd].Get("fullbar_tw/avrg_resol_cont")
    h.SetTitle("%s:%s"%(h.GetTitle(),mthd))
    h.SetMarkerStyle(ROOT.gROOT.ProcessLine("kFullStar"))#ROOT.gROOT.ProcessLine
    h.SetMarkerSize(2)
    h.SetLineColor(ROOT.gROOT.ProcessLine("kBlack"))
    h.SetStats(0)
    h_avg_rsltn_cmb[mthd]=h


#! Draw histograms
ROOT.gStyle.SetOptStat(0)
#ROOT.gStyle.SetErrorX(0.001)
c=rootnotes.canvas(size=(1000,400))
c.Divide(3,1)
l={}
for mthd in ['MLE','CSQ']:
    if mthd=='MLE':c.cd(1)
    else:          c.cd(2)
    l[mthd]=ROOT.TLegend(0.1,0.1,0.9,0.9)
    for i,h in enumerate(h_avg_rsltnl[mthd]):
        if i==0:
            h.SetMinimum(54)
            h.SetMaximum(64)
            h.Draw()
        else:
            h.Draw("same")
        l[mthd].AddEntry(h,run_namel[mthd][i])
    h_avg_rsltn_cmb[mthd].Draw("same")
    l[mthd].AddEntry(h_avg_rsltn_cmb[mthd],run_name_cmb[mthd])
c.cd(3)
l['MLE'].SetTextSize(0.05)
l['MLE'].Draw()
c
```
