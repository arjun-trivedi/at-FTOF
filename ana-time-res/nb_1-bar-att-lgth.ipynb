```python
from __future__ import division
import rootnotes
import re

#! datadir=location of FTOF official data (data from Gleb's analysis program)
datadir=os.path.join(os.environ['HOME'],"Downloads","ftof_official")

def MyFn(s):
    pos=s.find("cm")
    length=s[pos-3:pos]
    if length=="": #for bar length in 2 digits i.e 30cm to 94cm
        length=s[pos-2:pos]
    length=re.sub("[^0-9]", "", length)
    return int(length)
```

```python
#! Define al_dict
def get_al_dict():
    d={('pmt1','no'):0,('pmt2','no'):0,('pmt3','no'):0,
       ('pmt3','no'):0,('pmt5','no'):0,('pmt6','no'):0,
       ('pmt1','co'):0,('pmt2','co'):0,('pmt3','co'):0,
       ('pmt4','co'):0,('pmt5','co'):0,('pmt6','co'):0,
       'bar1':0,'bar2':0,'bar3':0,
       'bar4':0,'bar5':0,'bar6':0,
     }
    return d
```

```python
d=get_al_dict()
print d
```

```python
#! Loop through datadir/<length> and calculate att-lgth for each bar 

#! 1. Obtain att_length_123456.dat, att_length_321654.dat and fill al_dict 
for d in os.listdir(datadir):
    if os.path.isdir(os.path.join(datadir,d)):
        print "dir=",d
        #if d != "17": continue
        al=get_al_dict()
        setdir=os.path.join(datadir,d)
        for f in os.listdir(setdir):
            if "att_length_123456" in f:
                f_nrm_ordr=open(os.path.join(datadir,setdir,f)).readlines()
            if "att_length_321654" in f:
                f_cmp_ordr=open(os.path.join(datadir,setdir,f)).readlines()
        
        #! Create dict to obtain i_co from i_no
        idx_co={0:4,1:5,2:2,3:3,4:0,5:1,6:10,7:11,8:8,9:9,10:6,11:7}
        for i in range(12):
            i_no=i
            i_co=idx_co[i_no]
            al['pmt%d'%(i+1),'no']=round(float(f_nrm_ordr[i_no]),2)
            al['pmt%d'%(i+1),'co']=round(float(f_cmp_ordr[i_co]),2)
            #print 'pmt%d'%(i+1),'no',al['pmt%d'%(i+1),'no']
            #print 'pmt%d'%(i+1),'co',al['pmt%d'%(i+1),'co']
            print 'pmt%d'%(i+1),'avg',round((al['pmt%d'%(i+1),'no']+al['pmt%d'%(i+1),'co'])/2,2)
        #! Now compute att-lgth
        for nbar in range(1,7):
            lpmt=nbar*2-1
            rpmt=nbar*2
            print al['pmt%d'%lpmt,'no'],al['pmt%d'%rpmt,'no'],al['pmt%d'%lpmt,'co'],al['pmt%d'%rpmt,'co']
            al['bar%d'%nbar]=(1/2)*( 
                     (al['pmt%d'%lpmt,'no']+al['pmt%d'%rpmt,'no'])/2+
                     (al['pmt%d'%lpmt,'co']+al['pmt%d'%rpmt,'co'])/2 
                    )
        #! Now write att-lgth
        f_att_lgth=open("%s/att_length.txt"%setdir, "w")
        for nbar in range(1,7):
            f_att_lgth.write("%s\n"%al['bar%d'%nbar])
        f_att_lgth.close()              
```

```python
#! Loop through datadir/<length> and get all resolution data to be plotted.
#! Data=res from
#! 1. each 3-bar cmbn from
#!    + norm-order
#!    + comp-prder
#! 2. <3-bar cmbn> from
#!    + norm-order
#!    + comp-order
#!    + norm-order+comp-order
#! 3. 1-bar
#!    + mtrx-mthd
#!    + mean-mthd
#! 4. <1-bar-mtrx-mthd>

#! Create a dict. for res
#! res={length:{1:{(cmbn,ordr):},2:{ordr:},3:{(brnm,mthd):}},4:<1-bar-mtrx-mthd>}
#!    where:
#!    + cmbn='1_2_3','2_3_4','3_4_5','4_5_6','1_3_5','2_4_6'
#!    + ordr=norm,comp,norm-comp
#!    + brnm=1,2,3,4,5,6
#!    + mthd=mtrx,mean

def get_res_dict():
    d={1:
        {
          ('1_2_3','norm'):0,('2_3_4','norm'):0,
          ('3_4_5','norm'):0,('4_5_6','norm'):0,
          ('1_3_5','norm'):0,('2_4_6','norm'):0,
          ('1_2_3','comp'):0,('2_3_4','comp'):0,
          ('3_4_5','comp'):0,('4_5_6','comp'):0,
          ('1_3_5','comp'):0,('2_4_6','comp'):0
        },
      2:
        {'norm':0,'comp':0,'norm-comp':0},
      3:
        {
          ('bar1','mtrx'):0,('bar2','mtrx'):0,('bar3','mtrx'):0,
          ('bar4','mtrx'):0,('bar5','mtrx'):0,('bar6','mtrx'):0,
          ('bar1','mean'):0,('bar2','mean'):0,('bar3','mean'):0,
          ('bar4','mean'):0,('bar5','mean'):0,('bar6','mean'):0
        },
      4:0
     }
    return d
```

```python
#! Execution of above 
res={}
for d in os.listdir(datadir):
    if os.path.isdir(os.path.join(datadir,d)):
        print "dir=",d
        lgth=int(d)
        res[lgth]=get_res_dict()
        #print res[lgth]
        setdir=os.path.join(datadir,d)
        for f in os.listdir(setdir):
            if "123456" in f:f_nrm_ordr=ROOT.TFile(os.path.join(datadir,setdir,f))
            if "321654" in f:f_cmp_ordr=ROOT.TFile(os.path.join(datadir,setdir,f))
        f_gary_ana=open(os.path.join(datadir,d,"gary_ana","means.dat")).readlines()
        #print f_nrm_ordr,f_cmp_ordr,f_gary_ana
        
        #! get res[lgth][1][cmbn][norm]
        f=f_nrm_ordr
        h_avg_rsltnl=f.Get("fullbar_tw/avrg_resol_cont")
        res[lgth][1]['1_2_3','norm']=round(h_avg_rsltnl.GetBinContent(1),2)
        res[lgth][1]['2_3_4','norm']=round(h_avg_rsltnl.GetBinContent(2),2)
        res[lgth][1]['3_4_5','norm']=round(h_avg_rsltnl.GetBinContent(3),2)
        res[lgth][1]['4_5_6','norm']=round(h_avg_rsltnl.GetBinContent(4),2)
        res[lgth][1]['1_3_5','norm']=round(h_avg_rsltnl.GetBinContent(5),2)
        res[lgth][1]['2_4_6','norm']=round(h_avg_rsltnl.GetBinContent(6),2)
        #! get res[lgth][1][cmbn][comp]
        f=f_cmp_ordr
        h_avg_rsltnl=f.Get("fullbar_tw/avrg_resol_cont")
        res[lgth][1]['1_2_3','comp']=round(h_avg_rsltnl.GetBinContent(1),2)
        res[lgth][1]['2_3_4','comp']=round(h_avg_rsltnl.GetBinContent(2),2)
        res[lgth][1]['3_4_5','comp']=round(h_avg_rsltnl.GetBinContent(3),2)
        res[lgth][1]['4_5_6','comp']=round(h_avg_rsltnl.GetBinContent(4),2)
        res[lgth][1]['1_3_5','comp']=round(h_avg_rsltnl.GetBinContent(5),2)
        res[lgth][1]['2_4_6','comp']=round(h_avg_rsltnl.GetBinContent(6),2)
        
        #! get res[lgth][2][norm]
        res[lgth][2]['norm']=round(float(f_gary_ana[13]),2)
        #! get res[lgth][2][comp]
        res[lgth][2]['comp']=round(float(f_gary_ana[14]),2)
        #! get res[lgth][2][norm-comp]
        res[lgth][2]['norm-comp']=round(float(f_gary_ana[15]),2)
        
        #! get res[lgth][3][bar#][mtrx]
        for i in range(0,6):
            res[lgth][3]['bar%d'%(i+1),'mtrx']=round(float(f_gary_ana[i]),2)
        #! get res[lgth][3][bar#][mean]
        for i in range(6,12):
            res[lgth][3]['bar%d'%(i+1-6),'mean']=round(float(f_gary_ana[i]),2)
            
        #! get res[lgth][4]
        res[lgth][4]=round(float(f_gary_ana[12]),2)
        
        print res[lgth]
```

```python
#! Now make the plots

#! Plot 3-bar comb analysis results res (res[1][cmbn][ordr])
figsize_x,figsize_y=55,50
figtitle_fontsize=40

plottitle_fontsize=40
axislabel_fontsize=30
markersize=30
legend_size=40

fig=plt.figure(figsize=(figsize_x,figsize_y))
fig.suptitle("3-bar comb analysis results",fontsize=figtitle_fontsize)

subplot={'norm':1,'comp':2}
clrd={'1_2_3':'blue','2_3_4':'green','3_4_5':'red','4_5_6':'cyan','1_3_5':'magenta','2_4_6':'yellow'}

for ordr in ['norm','comp']:
    ax=plt.subplot(2,1,subplot[ordr])
    for cmbn in ['1_2_3','2_3_4','3_4_5','4_5_6','1_3_5','2_4_6']:
        x=[]
        y=[]
        for k in res.keys():
            x.append(k)
            y.append(res[k][1][cmbn,ordr])
        #ax.scatter(x,y,c=clrd[cmbn],label=cmbn,s=50)
        #! plot with 1ps stat-err (=2ps/2 where 2ps ~ width of estimated stat fluctuations)
        #! estimated from nb_40dr-2dc_no-Jul-12-13
        ax.errorbar(x,y,yerr=1,fmt='o',marker='o',ms=markersize,mfc=clrd[cmbn],mec=clrd[cmbn],label=cmbn) 
        ax.set_title(ordr,fontsize=plottitle_fontsize)
        ax.legend(loc='upper left',prop={'size':legend_size})
        #! Turn on ticks on right edge of plot too
        ax.tick_params(labelright=True)
        #! x axis aesthetics
        ax.set_xticks(sort(x))
        ax.set_xticklabels(sort(x),rotation='vertical',fontsize=axislabel_fontsize)
        ax.set_xlim(min(x)-20,max(x)+20)
        #! y axis aesthetics
        ymin=20
        ymax=110
        ystep=2
        yticks=range(ymin,ymax+ystep,ystep)
        ax.set_yticks(yticks)
        ax.set_yticklabels(yticks,fontsize=axislabel_fontsize)
        ax.set_ylim(ymin,ymax)
        #! Turn on grid
        ax.grid(linewidth=1)
fig.savefig("res_3-bar-cmb.png")
```
